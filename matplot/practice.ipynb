```python
data = {"package" : [11,3,43,4,54,4565.45,334,6343]}
df = pd.DataFrame(data)
df.head()
```

```python
df.shape
plt.plot(df["package"],color="red",marker="o")
plt.show()
```

```python
#histrogram
plt.hist(df["package"])
plt.show()
```

```python
#boxplot
plt.boxplot(df["package"])
plt.show()
```

```python
df["dept"]=['HR','IT','HR','IT','HR','IT','HR','IT']
df.head()
```

```python
#piechart
count = df["dept"].value_counts()

plt.pie(count)
plt.show
```

```python
plt.bar(count.index,count,color=['red','blue'])
plt.show()
```
