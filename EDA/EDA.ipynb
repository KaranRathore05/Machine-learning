```python
import seaborn as sns
import pandas as pd
```

```python

tt=pd.read_csv('titanic.csv')
tt
```

1.SCATTERPLOT(NUM-NUM)

```python
sns.scatterplot(x=tt['Fare'],y= tt['Age'])
```

2.Barplot (numerical - categorical)

```python
sns.barplot(x=tt['Pclass'],y=tt['Age'],hue=tt['Sex'])
```

3.BOX PLAOT(NUM-CAT)

```python
sns.boxplot(x=tt['Sex'],y=tt['Age'],hue=tt['Survived'])
```

4.DISPLOT(NUMERICAL - CATEGORICAL)

```python
sns.displot(y=tt['Survived']==0,x=tt['Age'])
sns.displot(y=tt['Survived']==1,x=tt['Age'])
```

HEATMAP(CAT-CAT)

```python
sns.heatmap(pd.crosstab(tt['Pclass'],tt['Survived']))
```

```python
tt.groupby('Pclass')['Survived'].mean() * 100
```
